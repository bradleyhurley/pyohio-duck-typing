```python
class Dog:
    def __init__(self, paws: int, puppy: bool, **kwargs: str):
        self.paws = paws
        self.is_puppy = puppy
        self.kwargs = kwargs
        
    def get_dog_breed(self) -> str:
        return self.kwargs.get('breed')
 
dog = Dog(paws=4, puppy=False, **{'breed':'boxer', 'color':'brown'})
print(f"{dog.get_dog_breed()}")
```

```python
from typing import Iterator
def fib(n: int) -> Iterator[int]:
    a, b = 0, 1
    while a < n:
        yield a
    a, b = b, a+b
```
